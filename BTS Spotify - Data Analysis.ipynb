# Clustering

```python
# merge it together

df_cluster = pd.DataFrame()
df_cluster = df[['track_id','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
df_cluster = df_cluster.set_index('track_id')
```

```python
from sklearn.cluster import KMeans
```

```python
# Calculate sum of squared distances
ssd = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df_cluster)
    ssd.append(km.inertia_)
```

```python
# Plot sum of squared distances / elbow method
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
model = kmeans.fit(df_cluster)
```

```python
pred = model.labels_
df_cluster['cluster'] = pred
df_cluster
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
# merge it together

df_pca = pd.DataFrame()
df_pca = df[['track_id','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
df_pca = df_pca.set_index('track_id')
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df_pca.values)
df_pca['pca-one'] = pca_result[:,0]
df_pca['pca-two'] = pca_result[:,1] 
df_pca['pca-three'] = pca_result[:,2]
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
df_pca['cluster'] = pred
df_pca
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(
    x="pca-one", y="pca-two",
    hue="cluster",
    data=df_pca,
    legend="full",
    alpha=0.3
)
plt.title("2D PCA Visualization with 3 Components", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(
    x="pca-one", y="pca-three",
    hue="cluster",
    data=df_pca,
    legend="full",
    alpha=0.3
)
plt.title("2D PCA Visualization with 3 Components", fontsize=16)
plt.show()
```

```python
ax = plt.figure(figsize=(16,10)).gca(projection='3d')
sctt = ax.scatter(
    xs = df_pca["pca-one"], 
    ys = df_pca["pca-two"], 
    zs = df_pca["pca-three"], 
    c = df_pca["cluster"]
)

ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')

plt.title("3D PCA Visualization with 3 Components", fontsize=16)

cb = plt.colorbar(sctt, ax=ax, shrink=0.3, aspect=4, pad=0.1)
cb.set_ticks([0,1,2,3])

plt.show()
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(df_tsne)
```

```python
df_tsne['cluster'] = pred
df_tsne
```

```python
df_tsne['tsne-2d-one'] = tsne_results[:,0]
df_tsne['tsne-2d-two'] = tsne_results[:,1]
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="cluster",
    palette=sns.color_palette("hls", 4),
    data=df_tsne,
    legend="full",
    alpha=0.3
)

plt.title("t-SNE 2D Visualization", fontsize=16)
```

```python
df_ref = pd.DataFrame(df[['track_id','name']])
df_ref = df_ref.set_index('track_id')
df_ref
```

```python
df_tsne = df_tsne.join(df_ref)
df_tsne
```

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']), size="small")
```

```python
df_tsne2 = pd.DataFrame(df_tsne[df_tsne['cluster']== 2])

plt.figure(figsize=(16,10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="cluster",
    palette=dict({2:"cornflowerblue"}),
    data=df_tsne2,
    legend="full",
    alpha=0.3,
)

label_point(df_tsne2['tsne-2d-one'], df_tsne2['tsne-2d-two'], df_tsne2.name, plt.gca())

plt.title("Cluster 2 Components", fontsize=16)
```
